```python
offline_idx = pd.read_excel(r"C:\Users\user\Downloads\offline_micro.xlsx", index_col=0)
```

```python
offline_idx = offline_idx.transpose()
offline_idx = offline_idx.set_index(pd.date_range(start='2019-02-28', end='2024-02-29', freq='M'))
```

```python
offline_idx = (offline_idx.pct_change()+1).fillna(1).cumprod().reindex(pd.date_range(start='2019-02-01', end='2024-02-29', freq='D'), method='bfill')
```

```python
import re

offline_idx.columns[4][:[m.start() for m in re.finditer(',', offline_idx.columns[4])][-1]]
```

```python
offline_idx.columns = list(map(lambda x: x[:[m.start() for m in re.finditer(',', x)][-1]], offline_idx.columns))
```
